```python
import pandas as pd

import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default = "plotly_white"
```

```python
data = pd.read_csv("sampledata.csv", encoding='latin-1')
data
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
#Converting Date Columns
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
data.info()
```

```python
data.head()
```

```python
data['Order month'] = data['Order Date'].dt.month
data['Order year'] = data['Order Date'].dt.year
data['Order Day o Week'] = data['Order Date'].dt.dayofweek
```

```python
data.head()
```

```python
#Monthly Sales Analysis
```

```python
sale_by_month=data.groupby('Order month')['Sales'].sum().reset_index()
sale_by_month
```

```python
fig = px.line(sale_by_month, x='Order month', y='Sales', title='Monthly Sales Analysis')
fig.show()
```

```python
data.head()
```

```python
sale_by_category=data.groupby('Category')['Sales'].sum().reset_index()
sale_by_category
```

```python
fig = px.pie(sale_by_category, values='Sales', names='Category', hole=0.1, color_discrete_sequence=colors.qualitative.Pastel)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Sales Analysis by Category', title_font=dict(size=24))
fig.show()
```

```python
sale_by_subcategory=data.groupby('Sub-Category')['Sales'].sum().reset_index()
sale_by_subcategory
```

```python
fig = px.bar(sale_by_subcategory, x='Sub-Category', y='Sales', title='Sales Analysis by Sub Category')
fig.show()
```

```python
# Sales and Profit for Customer Segment
```

```python
data.head()
```

```python
#sales_profit_by_segment = data.groupby('Segment').agg({'Sales','sum','Profit','sum'}).reset_index()
sales_profit_by_segment = data.groupby('Segment').agg(
    {
        'Sales': 'sum',
        'Profit': 'sum'
    }
).reset_index()
colour_palatee = colors.qualitative.Pastel

fig = go.Figure()
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                     y=sales_profit_by_segment['Sales'],
                     name='Sales',
                     marker_color = colour_palatee[0]))


fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                     y=sales_profit_by_segment['Profit'],
                     name='Profit',
                     marker_color = colour_palatee[1]))

fig.update_layout(title='Sales and Profit Analysis by Customer Segment',xaxis_title='Customer Segment',yaxis_title='Amount')
fig.show()
```

```python
# Sales to Profit Ratio
```

```python
sales_profit_by_segment = data.groupby('Segment').agg(
    {
        'Sales': 'sum',
        'Profit': 'sum'
    }
).reset_index()

sales_profit_by_segment['Sales_to_Profit_Ratio']=sales_profit_by_segment['Sales']/sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment', 'Sales_to_Profit_Ratio']])
```
