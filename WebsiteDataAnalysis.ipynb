```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

df = pd.read_csv("data-export (1).csv")
df.head()
```

```python
df.columns = df.iloc[0]
df = df.drop(index=0).reset_index(drop=True)
df.columns = ["Channel group", "DateHour", "Users", "Sessions", "Engagged Sessions",
              "Average engagement time per session", "Engaged sessions per user",
              "Events per session", "Engagement rate", "Event count"]
df.head()
```

```python
df.info()
```

```python
df["DateHour"]=pd.to_datetime(df["DateHour"],format="%Y%m%d%H",errors='coerce')
df.head()
```

```python
df.info()
```

```python
numeric_cols = df.columns.drop(["Channel group","DateHour"])
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors= 'coerce')
df["Hour"]=df["DateHour"].dt.hour
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sb.set(style="whitegrid")
```

```python
plt.figure(figsize=(10,5))
df.groupby("DateHour")[["Sessions","Users"]].sum().plot(ax=plt.gca())
plt.title("Sessions and Users over Time")
plt.xlabel("DateHours")
plt.ylabel("Count")
plt.show()
```

```python
#Total Users by Channel
```

```python
plt.figure(figsize=(8,6))
sb.barplot(data=df, x="Channel group", y="Users", estimator=np.sum, palette="viridis")
plt.title("Total Users b Channel")
plt.xticks(rotation=45)
plt.show()
```

```python
df.head()
```

```python
#Average Engagement Time by Channel
```

```python
plt.figure(figsize=(8,5))
sb.barplot(data=df, x="Channel group", y="Average engagement time per session", estimator=np.mean, palette="magma")
plt.title("Average Engagement Time by Channel")
plt.xticks(rotation=45)
plt.show()
```

```python
#Engagement Rate Distribution channel
```

```python
plt.figure(figsize=(8,5))
sb.boxplot(data=df, x="Channel group", y="Engagement rate", palette="coolwarm")
plt.title("Engagement Rate Distribution Channel")
plt.xticks(rotation=45)
plt.show()
```

```python
#Engagged vs non-Engagged Sessions
```

```python
df.head()
```

```python
sessions_df = df.groupby("Channel group")[["Sessions","Engagged Sessions"]].sum().reset_index()
sessions_df["Non-Engaged"]=sessions_df["Sessions"]-sessions_df["Engagged Sessions"]
session_df_melted = sessions_df.melt(id_vars="Channel group",value_vars=["Engagged Sessions","Non-Engaged"])


plt.figure(figsize=(8,5))
sb.barplot(data=session_df_melted, x="Channel group", y="value", hue="variable")
plt.title("Engagged vs non-Engagged Sessions")
plt.xticks(rotation=45)
plt.show()
```

```python
#Traffic by Hour and Channel
```

```python
heatmap_data = df.groupby(["Hour","Channel group"])["Sessions"].sum().unstack().fillna(0)

plt.figure(figsize=(8,5))
sb.heatmap(heatmap_data, cmap ="YlGnBu", linewidths=.5, annot=True, fmt='.0f')
plt.title("Traffic by Hour and Channel")
plt.xlabel("Channel group")
plt.ylabel("Hour of Day")
plt.xticks(rotation=45)
plt.show()
```

```python
#Engagement rates vs session over time
```

```python
df_plot= df.groupby("DateHour")[["Engagement rate","Sessions"]].mean().reset_index()

plt.figure(figsize=(8,5))
plt.plot(df_plot["DateHour"],df_plot["Engagement rate"], label="Engagement rate", color="green")
plt.plot(df_plot["DateHour"],df_plot["Engagement rate"], label="Sessions", color="skyblue")
plt.title("Engagement rates vs session over time")
plt.xlabel("DateHour")
plt.grid(True)
plt.legend()
plt.show()
```
